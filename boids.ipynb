```python
import  matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv("ppt/dt0.9//velocity.txt",names=['v'])
data.shape
y = np.array(data['v'])
x= np.array([i for i in range(len(y))])
x.shape
```

```python
# 绘制直线
plt.figure()
# plt.plot(x,y1)
# 曲线与直线绘制一块
# num=3表示图片上方标题 变为figure3，figsize=(长，宽)设置figure大小
plt.figure(num=3,figsize=(8,5))

# 设置x轴范围
plt.xlim((1,len(y)))
# 设置轴y范围
plt.ylim((1,50))
# 设置坐标轴含义
# 注：英文直接写，中文需要后面加上fontproperties属性
plt.xlabel('Steps')
plt.ylabel('Same Velocity')
# -1到2区间，5个点，4个区间，平均分：[-1.,-0.25,0.5,1.25,2.]


# 设置legend图例

plt.plot(x,y) # 可添加label属性，只不过如果这里添加了，下面legend再添加，下面的就会覆盖此处的！
# 红色虚线直线宽度默认1.0

# plt.legend(handles=[l1,l2],prop={'family':'SimHei','size':15},loc='lower right',labels=['直线','曲线'])
plt.savefig("50.jpg")
plt.show()
```

```python
data = pd.read_csv("swarm.txt",names=['n'])
y = np.array(data['n'])
x= np.array([i for i in range(len(y))])

plt.figure()
plt.figure(num=3,figsize=(8,5))
# 设置x轴范围
plt.xlim((1,len(y)))
# 设置轴y范围
plt.ylim((1,10))
# 设置坐标轴含义
# 注：英文直接写，中文需要后面加上fontproperties属性
plt.xlabel('Steps')
plt.ylabel('Group')


plt.plot(x,y) # 可添加label属性，只不过如果这里添加了，下面legend再添加，下面的就会覆盖此处的！

plt.savefig("50_n.jpg")
plt.show()
```
